```python
# Drop columns where all values are zero
columns_all_zero = [col for col in df.columns if (df[col] == 0).all()]

# Drop these columns from the DataFrame
df = df.drop(columns=columns_all_zero)

# Print the columns that were dropped
print("Dropped columns with all zero values:")
print(columns_all_zero)

# Optionally, check remaining columns
print("Remaining columns:")
print(df.columns)
```

```python
# Drop rows where all values are zero
rows_all_zero = df.index[(df == 0).all(axis=1)]

# Drop these rows from the DataFrame
df = df.drop(index=rows_all_zero)

# Print the indices of the rows that were dropped
print("Dropped rows with all zero values:")
print(rows_all_zero.tolist())

# Optionally, check the remaining DataFrame
print("Remaining rows and columns:")
print(df.shape)  # Prints the number of remaining rows and columns
```

```python
# Columns to add noise
columns_with_noise = [
    'Flow Duration', 'Total Fwd Packets', 'Bwd Packet Length Mean', 'Flow Bytes/s',
    'Flow Packets/s', 'Flow IAT Mean', 'Packet Length Mean', 'RST Flag Count', 'Down/Up Ratio',
    'ECE Flag Count', 'Average Packet Size', 'Active Mean', 'Idle Mean', 'attack'
]

# Set the noise factor (adjust this for desired impact)
noise_factor = 0.1

# Adding noise to the selected columns
for col in columns_with_noise:
    std_dev = df_dropped[col].std()  # Calculate standard deviation of the column
    noise = np.random.normal(0, noise_factor * std_dev, df_dropped.shape[0])  # Generate noise
    df_dropped[col] += noise  # Add noise to the column

# Check the updated DataFrame
print("Added noise to columns:", columns_with_noise)
print(df_dropped[columns_with_noise].head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, f1_score, accuracy_score

# Load the cleaned dataset
print("Loading dataset...")
df = pd.read_csv('CICIDS2017_18_cleaned.csv')
print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")

# Separate features and target variable
X = df.drop('Label', axis=1)  # Replace 'Label' with the actual name of your target column
y = df['Label']  # Keep y as a single-label array
y_encoded = pd.get_dummies(y)  # One-hot encode for multi-class metrics and models that support it

print(f"Target variable has been encoded into {y_encoded.shape[1]} classes.")

# Normalize the features
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
print(f"Training set size: {X_train.shape[0]} samples; Testing set size: {X_test.shape[0]} samples.")

# Define individual models
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
svm = SVC(probability=True, kernel='rbf', random_state=42)  # SVM requires y to be 1D
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

# Train and evaluate individual models
models = [('Decision Tree', dt), ('Random Forest', rf), ('KNN', knn), ('SVM', svm), ('XGBoost', xgb)]

print("Training individual models...")
for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {accuracy:.4f}")

# Define Soft Voting Classifier
print("Combining models into a Soft Voting Classifier...")
voting_clf = VotingClassifier(estimators=models, voting='soft')
voting_clf.fit(X_train, y_train)

# Evaluate the Soft Voting Classifier
print("\nEvaluating the Soft Voting Classifier...")
y_pred_voting = voting_clf.predict(X_test)
voting_accuracy = accuracy_score(y_test, y_pred_voting)
print(f"Soft Voting Classifier Accuracy: {voting_accuracy:.4f}")

# Classification report with actual labels
print("Classification report for Soft Voting Classifier...")
print(classification_report(y_test, y_pred_voting, target_names=y_encoded.columns))

# Calculate and print F1 Score
f1 = f1_score(y_test, y_pred_voting, average='weighted')
print(f'Soft Voting Classifier F1 Score: {f1:.4f}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, Flatten, MaxPooling1D
from sklearn.metrics import classification_report, f1_score

# Load the cleaned dataset
print("Loading dataset...")
df = pd.read_csv('CICIDS2017_18_cleaned.csv')
print(f"Dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")

# Separate features and target variable
X = df.drop('Label', axis=1)  # Replace 'Label' with the actual name of your target column
y = df['Label']

# One-hot encode the target variable for multi-label classification
print("One-hot encoding the target variable...")
y_encoded = pd.get_dummies(y)
print(f"Target variable has been encoded into {y_encoded.shape[1]} classes.")

# Normalize the features
print("Normalizing features...")
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Reshape X for CNN input (samples, timesteps, features)
print("Reshaping input for CNN...")
X = X.reshape(X.shape[0], X.shape[1], 1)

# Train-Test Split
print("Splitting the dataset into training and testing sets...")
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)
print(f"Training set size: {X_train.shape[0]} samples; Testing set size: {X_test.shape[0]} samples.")

# Define the CNN + MLP Model
print("Defining the CNN + MLP model...")
model = Sequential()
model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(MaxPooling1D(pool_size=2))

model.add(Conv1D(64, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))

model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Regularization
model.add(Dense(y_train.shape[1], activation='softmax'))

# Compile the model
print("Compiling the model...")
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the Model
print("Training the model...")
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.3, verbose=1)

# Get training accuracy and loss
train_loss = history.history['loss'][-1]  # Last training loss
train_accuracy = history.history['accuracy'][-1]  # Last training accuracy
print(f'Training Loss: {train_loss:.4f}, \nTraining Accuracy: {train_accuracy:.4f}')

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss:.4f}, \nTest Accuracy: {test_accuracy:.4f}')

# Make Predictions
print("Making predictions on the test set...")
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)  # For multi-class classification
y_true_classes = y_test.values.argmax(axis=1)

# Reverse one-hot encoding to get actual labels
actual_labels = y.unique()  # Get the unique original labels
y_pred_actual = actual_labels[y_pred_classes]  # Map predicted classes back to original labels
y_true_actual = actual_labels[y_true_classes]  # Map true classes back to original labels

# Classification report with actual labels
print("Generating classification report...")
print(classification_report(y_true_actual, y_pred_actual, target_names=actual_labels))

# Calculate and print F1 Score
f1 = f1_score(y_true_actual, y_pred_actual, average='weighted')
print(f'F1 Score: {f1:.4f}')

# Save the Model
model_save_path = r'C:\Users\HP\B.Tech\3rd year\5th sem\DL\Project Work\Project\intrusion_detection_model_cnn_mlp.h5'
model.save(model_save_path)
print(f"Saving the trained model to {model_save_path}...")
```

```python
import matplotlib.pyplot as plt

# Plotting function for accuracy and loss
def plot_history(history):
    print("Plotting training history...")
    plt.figure(figsize=(14, 5))
    
    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='upper left')
    
    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper left')
    
    plt.show()

# Call the plotting function
plot_history(history)
```
